```python
import pygame
pygame.init()
clock=pygame.time.Clock()
speed=60#fps
x=100
y=100
r=6
dx=2
dy=2
paddle_speed=5
paddle_w=6
paddle_h=40
display_width=500
display_height=300
paddle_x1=5
paddle_y1=5
paddle_x2=display_width-paddle_w-5
paddle_y2=5
display=pygame.display.set_mode((display_width,display_height))
pygame.display.set_caption("x")

def hit_sides():
    if y-r<0:
        return True
    if y+r>display_height:
        return True
    return False

def hit_paddle1():
    if x-r<paddle_x1+paddle_w and x-r>5 and y<paddle_y1+paddle_h and y>paddle_y1:
        return True
    return False

def hit_paddle2():
    if x+r>paddle_x2 and x+r<display_width-5 and y<paddle_y2+paddle_h and y>paddle_y2:
        return True
    return False

def hit_back():
    if x+r>display_width:
        return True
    return False

def game_over():
    x=True
    display.fill((0,0,0))
    title_font=pygame.font.Font(None,36)
    instruction_font=pygame.font.Font(None,24)
    prompt=title_font.render("GAME OVER",True,(255,0,0))
    prompt_disp=prompt.get_rect(center=(int(display_width/2),int(display_height/2)))
    display.blit(prompt,prompt_disp)
    prompt_re=title_font.render("Press r to Restart",True,(0,255,0))
    prompt_re_disp=prompt.get_rect(center=(int(display_width/2.2),int(display_height/1.5)))
    display.blit(prompt_re,prompt_re_disp)
    pygame.display.flip()
    while x:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                pygame.quit()
            if event.type==pygame.KEYDOWN:
                if event.key==pygame.K_r:
                    x=False

while True:
    clock.tick(speed)
    key=pygame.key.get_pressed()
    if key[pygame.K_s]:
        if paddle_y1+paddle_h+10<=display_height:
            paddle_y1+=paddle_speed
    if key[pygame.K_w]:
        if paddle_y1-10>=0:
            paddle_y1-=paddle_speed
    if key[pygame.K_DOWN]:
        if paddle_y2+paddle_h+10<=display_height:
            paddle_y2+=paddle_speed
    if key[pygame.K_UP]:
        if paddle_y2-10>=0:
            paddle_y2-=paddle_speed        
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            pygame.quit()
    display.fill((0,0,0))        
    x+=dx
    y+=dy
    pygame.draw.rect(display,(255,255,255),(paddle_x1,paddle_y1,paddle_w,paddle_h))
    pygame.draw.rect(display,(255,255,255),(paddle_x2,paddle_y2,paddle_w,paddle_h))
    pygame.draw.circle(display,(255,255,255),(x,y),r)#display,color,startion_position,radious
    if x<r or x+r>display_width:
        game_over()
    if hit_paddle1() or hit_paddle2():
        dx*=-1
    if hit_sides():
        dy*=-1
    pygame.display.update()
```
